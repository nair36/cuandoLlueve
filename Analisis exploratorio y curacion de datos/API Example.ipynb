```python
mydb = mysql.connector.connect(
  host=config.DB_HOST,
  user=config.DB_USER,
  passwd=config.DB_PASSWORD,
  database=config.DB_DATABASE
)

insertCursor = mydb.cursor()

baseDate = datetime.datetime(1970,1,1)

start = datetime.datetime(2017, 1, 1, 12, 0, 0)
end = datetime.datetime(2018, 1, 1)
delta = end - start
for i in range(delta.days + 1):
    currentDate = start + datetime.timedelta(days=i)
    currentDateInUnuxTime = int((currentDate - baseDate).total_seconds())
    print(str(currentDate))
   
    replacedUrl = baseUrl.replace("<key>", config.API_KEY).replace("<longitude>", str(config.TEODELINA_LONG)).replace("<latitude>", str(config.TEODELINA_LAT)).replace("<date>", str(currentDateInUnuxTime))
    response = requests.get(url = replacedUrl)
    data = response.json()
    
    sqlInsertQuery = 'INSERT INTO forecastHistory(data) VALUES ("<dataValue>")'
    
    sqlReplacedQuery = sqlInsertQuery.replace("<dataValue>",str(data))
    result  = insertCursor.execute(sqlReplacedQuery)
    mydb.commit()
    
mydb.close()
```
